```python
#select random cells
cell_ids=np.arange(pdata.X.shape[0])
rnd_cells = np.random.permutation(cell_ids)
train_cells = rnd_cells[:int(pdata.X.shape[0]/2.)]
test_cells=rnd_cells[int(pdata.X.shape[0]/2.):]
```

```python
pdata_train=pdata[train_cells,:]
pdata_test=pdata[test_cells,:]

adata_train=adata[train_cells,:]
adata_test=adata[test_cells,:]
```

```python
bdata=adata[:,pdata.var_names]
bdata.smp['louvain_groups'] = pdata.smp['louvain_groups']
bdata_train=adata_train[:,pdata.var_names]
bdata_train.smp['louvain_groups'] = pdata_train.smp['louvain_groups']
bdata_test=adata_test[:,pdata.var_names]
bdata_test.smp['louvain_groups'] = pdata_test.smp['louvain_groups']
```

```python
cell_label='1'
logr_labels=[int(i==cell_label) for i in bdata_train.smp['louvain_groups']]
logr=LogisticRegression(penalty='l2', dual=False, tol=0.0001, C=1.0, fit_intercept=True, intercept_scaling=1, class_weight=None, random_state=None, solver='liblinear', max_iter=100, multi_class='ovr', verbose=0, warm_start=False, n_jobs=1)
logr.fit(bdata_train.X,logr_labels)

predicted_labels=logr.predict(bdata_test.X)
true_labels=np.array([int(i==cell_label) for i in bdata_test.smp['louvain_groups']])

#probability of error
print("P_error:",1-np.sum([true_labels[i]==predicted_labels[i] for i in range(len(true_labels))])/float(len(true_labels)))
```

## multicclass

```python
%%time
logr_labels=np.array(bdata_train.smp['louvain_groups'],dtype=int)
logr=LogisticRegression(penalty='l2', dual=False, tol=0.0001, C=1.0, fit_intercept=True, intercept_scaling=1, class_weight=None, random_state=None, solver='liblinear', max_iter=100, multi_class='ovr', verbose=0, warm_start=False, n_jobs=1)
logr.fit(bdata_train.X,logr_labels)

predicted_labels=logr.predict(bdata_test.X)
true_labels=np.array(bdata_test.smp['louvain_groups'],dtype=int)

#probability of error
print("P_error:",1-np.sum([true_labels[i]==predicted_labels[i] for i in range(len(true_labels))])/float(len(true_labels)))
```

```python
%%time
## no intercept
logr_labels=np.array(bdata_train.smp['louvain_groups'],dtype=int)
logr2=LogisticRegression(penalty='l2', dual=False, tol=0.0001, C=1.0, fit_intercept=False, intercept_scaling=1, class_weight=None, random_state=None, solver='liblinear', max_iter=100, multi_class='ovr', verbose=0, warm_start=False, n_jobs=1)
logr2.fit(bdata_train.X,logr_labels)

predicted_labels=logr2.predict(bdata_test.X)
true_labels=np.array(bdata_test.smp['louvain_groups'],dtype=int)

#probability of error
print("P_error:",1-np.sum([true_labels[i]==predicted_labels[i] for i in range(len(true_labels))])/float(len(true_labels)))
```
